```python
# 1. Create a list with 5 names and output the 2nd item
names = ["Alice", "Bob", "Charlie", "David", "Eve"]
print("Second item:", names[1])  

# 2. Change the value of the first item
names[0] = "Alex"
print("Updated list:", names) 

# 3. Add a sixth item
names.append("Frank")
print("List with sixth item:", names)  

# 4. Add "Bathel" as the 3rd item
names.insert(2, "Bathel")
print("List with Bathel:", names) 

# 5. Remove the 4th item
del names[3]
print("List after removal:", names)  

# 6. Negative indexing to print the last item
print("Last item:", names[-1]) 

# 7. New list with 7 items, print 3rd, 4th, and 5th items
new_list = ["Apple", "Banana", "Cherry", "Date", "Elderberry", "Fig", "Grape"]
print("Items 3-5:", new_list[2:5])  

# 8. List of countries and make a copy
countries = ["USA", "UK", "Canada", "Japan"]
countries_copy = countries.copy()
print("Copied list:", countries_copy) 

# 9. Loop through the list of countries
print("Countries:")
for country in countries:
    print(country)

# 10. List of animals sorted in ascending and descending order
animals = ["Lion", "Tiger", "Elephant", "Giraffe"]
ascending = sorted(animals)
descending = sorted(animals, reverse=True)
print("Ascending:", ascending)  
print("Descending:", descending) 

# 11. Output animal names with 'a' in them
animals_with_a = [animal for animal in animals if 'a' in animal.lower()]
print("Animals with 'a':", animals_with_a)  

# 12. Join two lists (first names and second names)
first_names = ["John", "Jane"]
last_names = ["Doe", "Smith"]
full_names = first_names + last_names
print("Joined lists:", full_names) 
```

```python
# 1. Output favorite phone brand
phones = ("samsung", "iphone", "tecno", "redmi")
print("Favorite brand:", phones[1]) 

# 2. Negative indexing to print 2nd last item
print("Second last item:", phones[-2])   

# 3. Update "iphone" to "itel" (Tuples are immutable, so convert to list)
phones_list = list(phones)
phones_list[1] = "itel"
phones = tuple(phones_list)
print("Updated tuple:", phones)  

# 4. Add "Huawei" (Tuples are immutable, so create a new tuple)
phones += ("Huawei",)
print("Tuple with Huawei:", phones)  

# 5. Loop through the tuple
print("Phone brands:")
for phone in phones:
    print(phone)

# 6. Remove first item (convert to list, modify, then back to tuple)
phones_list = list(phones)
phones_list.pop(0)
phones = tuple(phones_list)
print("After removing first item:", phones)   

# 7. Tuple of Ugandan cities
uganda_cities = tuple(["Kampala", "Entebbe", "Jinja", "Mbarara"])
print("Uganda cities:", uganda_cities)   

# 8. Unpack the tuple
city1, city2, city3, city4 = uganda_cities
print("Unpacked cities:", city1, city2, city3, city4)   

# 9. Range of indexes (2nd, 3rd, 4th cities)
print("Cities 2-4:", uganda_cities[1:4])  

# 10. Join two tuples
first_names = ("John", "Jane")
last_names = ("Doe", "Smith")
full_names = first_names + last_names
print("Joined tuples:", full_names)  

# 11. Multiply tuple of colors by 3
colors = ("red", "blue")
multiplied_colors = colors * 3
print("Multiplied colors:", multiplied_colors)  

# 12. Count occurrences of 8 in the tuple
thistuple = (1, 3, 7, 8, 7, 5, 4, 6, 8, 5)
count_8 = thistuple.count(8)
print("Number of 8s:", count_8)   
```

```python
# 1. Create a set of beverages
beverages = set(["Coffee", "Tea", "Juice"])
print("Beverages:", beverages)   

# 2. Add 2 more items
beverages.add("Water")
beverages.add("Milk")
print("Updated beverages:", beverages)  

# 3. Check if "microwave" is in the set
mySet = {"oven", "kettle", "microwave", "refrigerator"}
print("Is microwave present?", "microwave" in mySet) 

# 4. Remove "kettle"
mySet.remove("kettle")
print("Set after removal:", mySet)   

# 5. Loop through the set
print("Kitchen appliances:")
for item in mySet:
    print(item)

# 6. Add list elements to a set
set_items = {"apple", "banana", "cherry"}
list_items = ["date", "elderberry"]
set_items.update(list_items)
print("Updated set:", set_items)   

# 7. Join two sets (union)
ages = {25, 30}
first_names = {"John", "Jane"}
combined = ages.union(first_names)
print("Combined set:", combined)  
```

```python
# 1. Print shoe size
Shoes = {"brand": "Nick", "color": "black", "size": 40}
print("Shoe size:", Shoes["size"])   

# 2. Change "Nick" to "Adidas"
Shoes["brand"] = "Adidas"
print("Updated Shoes:", Shoes) 

# 3. Add "type": "sneakers"
Shoes["type"] = "sneakers"
print("Shoes with type:", Shoes)  

# 4. Return all keys
print("Keys:", list(Shoes.keys()))  

# 5. Return all values
print("Values:", list(Shoes.values()))   

# 6. Check if "size" exists
print("Does 'size' exist?", "size" in Shoes)  

# 7. Loop through the dictionary
print("Shoe details:")
for key, value in Shoes.items():
    print(f"{key}: {value}")

# 8. Remove "color"
Shoes.pop("color")
print("After removing color:", Shoes)  

# 9. Empty the dictionary
Shoes.clear()
print("Empty dictionary:", Shoes)  

# 10. Make a copy of a dictionary
original = {"name": "John", "age": 30}
copy = original.copy()
print("Copied dictionary:", copy)   

# 11. Nested dictionaries
nested = {
    "person1": {"name": "Alice", "age": 25},
    "person2": {"name": "Bob", "age": 30}
}
print("Nested dictionary:", nested)
```
